For this project, real-time webcam analysis was not implemented. The decision was made due to privacy considerations, as the use of a personal webcam was not preferred.

Instead, the project focuses on static image analysis using multiple test images, which still effectively demonstrates the capabilities of the DeepFace framework in predicting age, gender, race and emotion. This approach ensures that all core objectives of the assignment are successfully achieved while maintaining user privacy.

# 2. Import Libraries

```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import pandas as pd
import numpy as np
```

# 3. Upload One or More Images

```python
uploaded = files.upload()
image_paths = list(uploaded.keys())
print("Uploaded Images:", image_paths)
```

# 4. Display Uploaded Images

```python
for img_path in image_paths:
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title(img_path)
    plt.axis('off')
    plt.show()
```

# 5. Perform Face Analysis

```python
results_list = []

for img_path in image_paths:
    print("\nProcessing:", img_path)

    result = DeepFace.analyze(
        img_path=img_path,
        actions=['age','gender','race','emotion'],
        enforce_detection=False
    )

    age = result[0]['age']
    gender = result[0]['dominant_gender']
    race = result[0]['dominant_race']
    emotion = result[0]['dominant_emotion']

    print(f"Age: {age}, Gender: {gender}, Race: {race}, Emotion: {emotion}")

    results_list.append({
        "Image": img_path,
        "Age": age,
        "Gender": gender,
        "Race": race,
        "Emotion": emotion
    })
```

# 6. Results Table

```python
df = pd.DataFrame(results_list)
df
```

# 7. Visualize Predictions on Images

```python
for img_path in image_paths:
    img = cv2.imread(img_path)

    result = DeepFace.analyze(
        img_path=img_path,
        actions=['age','gender','race','emotion'],
        enforce_detection=False
    )

    age = result[0]['age']
    gender = result[0]['dominant_gender']
    race = result[0]['dominant_race']
    emotion = result[0]['dominant_emotion']

    text = f"Age:{age}, {gender}, {race}, {emotion}"

    cv2.putText(img, text, (10,30),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.6, (0,255,0), 2)

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(img_path)
    plt.axis('off')
    plt.show()
```
