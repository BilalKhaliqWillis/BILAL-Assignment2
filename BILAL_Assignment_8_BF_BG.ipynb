Task 1: Subsumption Function

```python
from nltk.featstruct import FeatStruct, unify

def subsumes(fstruct1, fstruct2):
    result = unify(fstruct1, fstruct2)
    return result == fstruct2

# Define feature structures
fs1 = FeatStruct(NAME="Jack Torrance", ROOM="237", LODGING="Overlook Hotel")
fs2 = FeatStruct(NAME="Jack Torrance", ROOM="237", LODGING="Overlook Hotel")
fs3 = FeatStruct(NAME="Jack Torrance", LODGING="Overlook Hotel")

print("fs1 subsumes fs2:", subsumes(fs1, fs2))
print("fs3 subsumes fs1:", subsumes(fs3, fs1))
print("fs1 subsumes fs3:", subsumes(fs1, fs3))
```

# Task 2: Agreement Grammar

```python
grammar = nltk.grammar.FeatureGrammar.fromstring('''
S -> NP[NUM=?n, PER=?p] VP[NUM=?n, PER=?p]

NP[NUM=sg, PER=1] -> 'I'
NP[NUM=sg, PER=2] -> 'you'
NP[NUM=sg, PER=3] -> 'he'
NP[NUM=pl, PER=1] -> 'we'

VP[NUM=?n, PER=?p] -> V[NUM=?n, PER=?p] PP

VP[NUM=?n, PER=?p] -> AUX[NUM=?n, PER=?p] VPROG PP
VP[NUM=?n, PER=?p] -> AUXHAVE[NUM=?n, PER=?p] VPASTPART PP
VP -> VPAST PP

PP -> 'to' 'the' 'store'

V[NUM=sg, PER=1] -> 'go'
V[NUM=sg, PER=2] -> 'go'
V[NUM=pl] -> 'go'
V[NUM=sg, PER=3] -> 'goes'

VPAST -> 'went'
VPROG -> 'going'
VPASTPART -> 'gone'

AUX[NUM=sg, PER=1] -> 'am'
AUX[NUM=sg, PER=3] -> 'is'
AUX[NUM=pl] -> 'are'
AUX[NUM=sg, PER=2] -> 'are'

AUXHAVE[NUM=sg, PER=3] -> 'has'
AUXHAVE -> 'have'
''')

parser = nltk.parse.FeatureEarleyChartParser(grammar)

sentences = [
    "I go to the store",
    "he is going to the store",
    "we have gone to the store",
    "you went to the store"
]

for sent in sentences:
    print("\nSentence:", sent)
    tokens = sent.split()
    for tree in parser.parse(tokens):
        print(tree)
```

# Task 3: Verb Subcategorization

```python
grammar2 = nltk.grammar.FeatureGrammar.fromstring('''
S -> NP VP

NP -> Det N
Det -> 'the'
N -> 'farmer' | 'cart' | 'sand'

VP -> V[TYPE=load] NP PP_with
VP -> V[TYPE=load] NP PP_into
VP -> V[TYPE=fill] NP PP_with
VP -> V[TYPE=dump] NP PP_into

PP_with -> 'with' NP
PP_into -> 'into' NP

V[TYPE=load] -> 'loaded'
V[TYPE=fill] -> 'filled'
V[TYPE=dump] -> 'dumped'
''')

parser2 = nltk.parse.FeatureEarleyChartParser(grammar2)

test_sentences = [
    "the farmer loaded the cart with sand",
    "the farmer loaded sand into the cart",
    "the farmer filled the cart with sand",
    "the farmer filled sand into the cart",
    "the farmer dumped the cart with sand",
    "the farmer dumped sand into the cart"
]

for sent in test_sentences:
    print("\nSentence:", sent)
    tokens = sent.split()
    trees = list(parser2.parse(tokens))
    if trees:
        for t in trees:
            print(t)
    else:
        print("No valid parse (Ungrammatical)")
```
