```python
import PyPDF2

with open("Business_Proposal.pdf", "rb") as pdf_file:
    reader = PyPDF2.PdfReader(pdf_file)
    page_two_text = reader.pages[1].extract_text()

print("Page 2 Text:\n")
print(page_two_text)
```

```python
# Cleaning and appending the text
cleaned_text = page_two_text.replace("AUTHORS:", "")

with open("contacts.txt", "a") as file:
    file.write("\n" + cleaned_text)

print("Text appended successfully!")
```

```python
# Displaying updated file
with open("contacts.txt", "r") as file:
    updated_content = file.read()

print("Updated contacts.txt:\n")
print(updated_content)
```

```python
import re

with open("Business_Proposal.pdf", "rb") as pdf_file:
    reader = PyPDF2.PdfReader(pdf_file)

    full_text = ""
    for page in reader.pages:
        full_text += page.extract_text()

email_pattern = r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+'
emails = re.findall(email_pattern, full_text)

print("Extracted Email Addresses:\n")
for email in emails:
    print(email)
```
