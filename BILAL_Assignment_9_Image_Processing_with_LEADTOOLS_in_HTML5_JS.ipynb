# Image Processing with LEADTOOLS
This Assignment creates image processing using:
- HTML5 Canvas
- JavaScript
- LEADTOOLS integration structure

# HTML

```python
%%writefile index.html
<!DOCTYPE html>
<html>
<head>
  <title>Image Processing Lab</title>
  <link rel="stylesheet" href="style.css">
</head>
<body>

<h2>Image Processing App</h2>

<input type="file" id="upload" accept="image/*">
<br><br>

<canvas id="canvas"></canvas>

<br><br>

<button onclick="applyGrayscale()">Grayscale</button>
<button onclick="resetImage()">Reset</button>

<br><br>

Brightness:
<input type="range" id="brightness" min="-100" max="100" value="0">

Contrast:
<input type="range" id="contrast" min="-100" max="100" value="0">

<br><br>

<button onclick="invertColors()">Invert Colors</button>

<script src="script.js"></script>

</body>
</html>
```

# CSS

```python
%%writefile style.css
body {
  text-align: center;
  font-family: Arial;
}

canvas {
  border: 1px solid black;
  max-width: 90%;
}
```

# JavaScript

```python
%%writefile script.js
window.onload = function () {

let canvas = document.getElementById('canvas');
let ctx = canvas.getContext('2d');
let originalImage = null;

// Upload Image
document.getElementById('upload').addEventListener('change', function(e) {
  let file = e.target.files[0];
  let img = new Image();

  img.onload = function() {
    canvas.width = img.width;
    canvas.height = img.height;
    ctx.drawImage(img, 0, 0);

    originalImage = ctx.getImageData(0, 0, canvas.width, canvas.height);
  };

  img.src = URL.createObjectURL(file);
});

// Grayscale
function applyGrayscale() {
  let imgData = ctx.getImageData(0, 0, canvas.width, canvas.height);
  let d = imgData.data;

  for (let i = 0; i < d.length; i += 4) {
    let avg = (d[i] + d[i+1] + d[i+2]) / 3;
    d[i] = d[i+1] = d[i+2] = avg;
  }

  ctx.putImageData(imgData, 0, 0);
}

// Brightness & Contrast
document.getElementById('brightness').addEventListener('input', adjust);
document.getElementById('contrast').addEventListener('input', adjust);

function adjust() {
  if (!originalImage) return;

  let brightness = parseInt(document.getElementById('brightness').value);
  let contrast = parseInt(document.getElementById('contrast').value);

  let imgData = new ImageData(
    new Uint8ClampedArray(originalImage.data),
    originalImage.width,
    originalImage.height
  );

  let d = imgData.data;

  let factor = (259 * (contrast + 255)) / (255 * (259 - contrast));

  for (let i = 0; i < d.length; i += 4) {
    d[i] = factor * (d[i] - 128) + 128 + brightness;
    d[i+1] = factor * (d[i+1] - 128) + 128 + brightness;
    d[i+2] = factor * (d[i+2] - 128) + 128 + brightness;
  }

  ctx.putImageData(imgData, 0, 0);
}

// Reset
function resetImage() {
  if (originalImage)
    ctx.putImageData(originalImage, 0, 0);
}

// Invert Colors
function invertColors() {
  let imgData = ctx.getImageData(0, 0, canvas.width, canvas.height);
  let d = imgData.data;

  for (let i = 0; i < d.length; i += 4) {
    d[i] = 255 - d[i];
    d[i+1] = 255 - d[i+1];
    d[i+2] = 255 - d[i+2];
  }

  ctx.putImageData(imgData, 0, 0);
}

// Make buttons work
window.applyGrayscale = applyGrayscale;
window.resetImage = resetImage;
window.invertColors = invertColors;

};
```
