# 2. Import Libraries

```python
import speech_recognition as sr
from gtts import gTTS
import os
import datetime
import webbrowser
import tempfile
import math
from IPython.display import Audio, display
```

# 3. Text-to-Speech Function

```python
def speak(text):
    tts = gTTS(text=text, lang='en')
    filename = "voice.mp3"
    tts.save(filename)

    display(Audio(filename, autoplay=True))
```

```python
speak("Assistant is ready")
```

# 4. Voice Input

```python
def take_command():
    recognizer = sr.Recognizer()
    try:
        with sr.Microphone() as source:
            print("Listening...")
            audio = recognizer.listen(source)
            command = recognizer.recognize_google(audio)
            print("You said:", command)
            return command.lower()
    except:
        return input("Type your command: ").lower()
```

# 5. Built-in Calculator (No API Needed)

```python
def calculate_expression(expression):
    try:
        expression = expression.replace("plus", "+").replace("minus", "-") \
                               .replace("times", "*").replace("multiplied by", "*") \
                               .replace("divided by", "/")
        return eval(expression)
    except:
        return "Error"
```

# 6. Command Processing

```python
def process_command(command):

    if "open google" in command:
        speak("Opening Google")
        webbrowser.open("https://www.google.com")

    elif "open youtube" in command:
        speak("Opening YouTube")
        webbrowser.open("https://www.youtube.com")

    elif "open wikipedia" in command:
        speak("Opening Wikipedia")
        webbrowser.open("https://www.wikipedia.org")

    elif "time" in command:
        now = datetime.datetime.now().strftime("%H:%M:%S")
        speak(f"The current time is {now}")

    elif "calculate" in command:
        query = command.replace("calculate", "")
        result = calculate_expression(query)
        speak(f"The answer is {result}")

    elif "search" in command:
        query = command.replace("search", "")
        speak(f"Searching for {query}")
        webbrowser.open(f"https://www.google.com/search?q={query}")

    elif "exit" in command or "bye" in command:
        speak("Goodbye!")
        return False

    else:
        speak("I did not understand that command.")

    return True
```

# 7. Run Assistant

```python
def run_assistant():
    speak("Hello! How can I help you?")
    running = True
    while running:
        command = take_command()
        if command:
            running = process_command(command)

run_assistant()
```
