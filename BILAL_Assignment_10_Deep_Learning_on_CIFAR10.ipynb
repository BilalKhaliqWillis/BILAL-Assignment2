```python
# Assignment 10- Deep Learning on CIFAR10
import tensorflow as tf
from tensorflow import keras

(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.cifar10.load_data()

print("Training data shape:", X_train_full.shape)
print("Test data shape:", X_test.shape)
```

```python
# Import Required Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
# Loading CIFAR-10 Dataset
(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.cifar10.load_data()

print("Training shape:", X_train_full.shape)
print("Test shape:", X_test.shape)
```

```python
# Train / Validation Split
X_valid = X_train_full[:5000]
y_valid = y_train_full[:5000]

X_train = X_train_full[5000:]
y_train = y_train_full[5000:]
```

```python
# Data Normalization
X_train = X_train.astype("float32") / 255.0
X_valid = X_valid.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
```

```python
# Flattening the Images
X_train_flat = X_train.reshape(len(X_train), -1)
X_valid_flat = X_valid.reshape(len(X_valid), -1)
X_test_flat = X_test.reshape(len(X_test), -1)

print("Flattened shape:", X_train_flat.shape)
```

```python
# Deep Neural Network (20 Layers)
# Callbacks -EarlyStopping, Checkpoint, TensorBoard
log_dir = "logs/dnn_" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

early_stop = keras.callbacks.EarlyStopping(
    patience=20,
    restore_best_weights=True
)

checkpoint = keras.callbacks.ModelCheckpoint(
    "best_dnn_model.h5",
    save_best_only=True
)

tensorboard = keras.callbacks.TensorBoard(log_dir=log_dir)
```

```python
# Building the 20-Layer DNN -Swish + He Initialization
model_dnn = keras.Sequential()
model_dnn.add(layers.Input(shape=(3072,)))

for _ in range(20):
    model_dnn.add(
        layers.Dense(
            100,
            activation="swish",
            kernel_initializer="he_normal"
        )
    )

model_dnn.add(layers.Dense(10, activation="softmax"))
```

```python
# Compiling the Model
model_dnn.compile(
    optimizer=keras.optimizers.Nadam(learning_rate=5e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model_dnn.summary()
```

```python
# Training the Model
history_dnn = model_dnn.fit(
    X_train_flat, y_train,
    validation_data=(X_valid_flat, y_valid),
    epochs=100,
    callbacks=[early_stop, checkpoint, tensorboard],
    batch_size=64
)
```

```python
# Evaluating on Test Set
test_loss, test_acc = model_dnn.evaluate(X_test_flat, y_test)
print("Test Accuracy:", test_acc)
```

```python
# Batch Normalization Experiment
# Building DNN with Batch Normalization
model_bn = keras.Sequential()
model_bn.add(layers.Input(shape=(3072,)))

for _ in range(20):
    model_bn.add(layers.Dense(100, kernel_initializer="he_normal"))
    model_bn.add(layers.BatchNormalization())
    model_bn.add(layers.Activation("swish"))

model_bn.add(layers.Dense(10, activation="softmax"))
```

```python
# Compiling BN Model
model_bn.compile(
    optimizer=keras.optimizers.Nadam(learning_rate=1e-4),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
# Training BN Model
history_bn = model_bn.fit(
    X_train_flat, y_train,
    validation_data=(X_valid_flat, y_valid),
    epochs=100,
    batch_size=64,
    callbacks=[early_stop, checkpoint, tensorboard]
)
```

```python
# Evaluating BN Model
test_loss_bn, test_acc_bn = model_bn.evaluate(X_test_flat, y_test)
print("BN Test Accuracy:", test_acc_bn)
```

Batch Normalization Analysis
•	Converges faster: Yes
•	Better accuracy: Usually a higher validation accuracy
•	Training speed: Slightly slower per epoch becauseof the BN computation
•	Overall: More stable and robust training


```python
# Self-Normalizing Network (SELU)
# Standardizing Inputs -needed for SELU
mean = X_train_flat.mean(axis=0)
std = X_train_flat.std(axis=0) + 1e-7

X_train_selu = (X_train_flat - mean) / std
X_valid_selu = (X_valid_flat - mean) / std
X_test_selu = (X_test_flat - mean) / std
```

```python
# Building SELU Model -LeCun Initialization
model_selu = keras.Sequential()
model_selu.add(layers.Input(shape=(3072,)))

for _ in range(20):
    model_selu.add(
        layers.Dense(
            100,
            activation="selu",
            kernel_initializer="lecun_normal"
        )
    )

model_selu.add(layers.Dense(10, activation="softmax"))
```

```python
# Compiling SELU Model
model_selu.compile(
    optimizer=keras.optimizers.Nadam(learning_rate=1e-4),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
# Training SELU Model
history_selu = model_selu.fit(
    X_train_selu, y_train,
    validation_data=(X_valid_selu, y_valid),
    epochs=100,
    batch_size=64,
    callbacks=[early_stop, checkpoint, tensorboard]
)
```

```python
# Evaluating SELU Model
test_loss_selu, test_acc_selu = model_selu.evaluate(X_test_selu, y_test)
print("SELU Test Accuracy:", test_acc_selu)
```
