```python
import nltk
from nltk.sem import Expression
from nltk.sem.evaluate import Model, Valuation, Assignment
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import string

nltk.download('punkt')
nltk.download('stopwords')
```

# Task 1: Logical Reasoning

```python
# Function to create logical statements
def represent_logic_statements():
    read_expr = Expression.fromstring

    stmt1 = read_expr('all x.(Cat(x) -> Friendly(x))')
    stmt2 = read_expr('exists x.(Dog(x) & Friendly(x))')

    return stmt1, stmt2
```

```python
# Defining valuation - facts about the world
val = Valuation([
    ('Cat', set(['cat1'])),
    ('Dog', set(['dog1'])),
    ('Mammal', set(['cat1'])),
    ('Friendly', set(['dog1']))
])

# Defining the domain -all objects
domain = set(['cat1', 'dog1'])

# Creating the model
model = Model(domain, val)
```

```python
# Corrected function to evaluate truth
def check_truth(statements, model):
    g = Assignment(model.domain)  # FIXED (important!)
    results = []

    for stmt in statements:
        result = model.evaluate(str(stmt), g)
        results.append(result)

    return results

# Running evaluation
statements = represent_logic_statements()
results = check_truth(statements, model)

for stmt, res in zip(statements, results):
    print(f"{stmt} => {res}")
```

```python
# Loading the corpus
with open('linguistic_data.txt', 'r') as f:
    corpus = f.read()

# Converting to sentences
sentences = sent_tokenize(corpus)
print(sentences)
```

```python
# Mapping function
def create_mapping(keys, sentences):
    mapping = {}
    for i, key in enumerate(keys):
        if i < len(sentences):
            mapping[key] = sentences[i]
    return mapping

keys = ['tokenization', 'stemming', 'lemmatization', 'vectorization']
mapping = create_mapping(keys, sentences)

mapping
```

# Task 3: Rule-Based Chatbot

```python
# Text preprocessing
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    stop_words = set(stopwords.words('english'))

    filtered = [
        word for word in tokens
        if word not in stop_words and word not in string.punctuation
    ]

    return filtered
```

```python
# Response function
def get_response(text, knowledge_base):
    tokens = preprocess_text(text)

    for word in tokens:
        if word in knowledge_base:
            return knowledge_base[word]

    return "Sorry, I don't understand."
```

```python

# Chatbot loop
def chat(knowledge_base):
    print("Chatbot: Hello! Type 'bye' to exit.")

    while True:
        user_input = input("You: ")

        if user_input.lower() == 'bye':
            print("Chatbot: Goodbye!")
            break

        response = get_response(user_input, knowledge_base)
        print("Chatbot:", response)

# Start the chatbot
chat(mapping)
```
