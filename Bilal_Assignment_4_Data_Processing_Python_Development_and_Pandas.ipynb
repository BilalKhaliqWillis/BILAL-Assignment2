```python
import pandas as pd
import numpy as np
from functools import reduce
import itertools

print("Bilal-Assignment 4\n")
```

```python
# Create a DataFrame with the given data for Name, Age, and City
try:
    data = {
        "Name": ["Alice", "Bob", "Charlie", "Diana"],
        "Age": [28, 34, 22, 45],
        "City": ["Toronto", "Montreal", "Vancouver", "Calgary"]
    }
    df = pd.DataFrame(data)
    # Ensure correct dtypes
    df['Name'] = df['Name'].astype(str)
    df['Age'] = df['Age'].astype(int)
    df['City'] = df['City'].astype(str)
    print("Task 1 - DataFrame created:")
    print(df)
    print(df.dtypes, "\n")
except Exception as e:
    print("Error in Task 1:", e)
```

```python
# Row and Column Manipulation:
# Drop 'City' column and verify
try:
    df_no_city = df.drop(columns=['City'])
    print("Task 2 - After dropping 'City':")
    print(df_no_city)
    # Verify
    expected_cols = set(['Name', 'Age'])
    actual_cols = set(df_no_city.columns)
    assert actual_cols == expected_cols, f"Columns mismatch: {actual_cols}"
    print("Verification passed: only 'Name' and 'Age' columns present.\n")
except Exception as e:
    print("Error in Task 2:", e)
```

```python
# Handling Null Values
# Create DataFrame with nulls and demonstrate handling methods
try:
    df_nulls = pd.DataFrame({
        "A": [1, np.nan, 3],
        "B": ["x", "y", None],
        "C": [10.5, np.nan, 30.2]
    })
    print("Task 3 - Original with nulls:")
    print(df_nulls, "\n")

    # Method 1: fill nulls (different filling strategies)
    df_filled = df_nulls.copy()
    df_filled['A'] = df_filled['A'].fillna(0)           # numeric -> 0
    df_filled['B'] = df_filled['B'].fillna("missing")   # string -> "missing"
    df_filled['C'] = df_filled['C'].fillna(df_filled['C'].mean())  # numeric -> mean
    print("Filled nulls (A->0, B->'missing', C->mean):")
    print(df_filled, "\n")

    # Method 2: drop rows with any nulls
    df_dropped = df_nulls.dropna()
    print("Dropped rows with any nulls:")
    print(df_dropped, "\n")

    # Method 3: forward-fill / back-fill
    df_ffill = df_nulls.fillna(method='ffill').fillna(method='bfill')
    print("Forward/backward filled:")
    print(df_ffill, "\n")
except Exception as e:
    print("Error in Task 3:", e)
```

```python
# GroupBy and Describe
# GroupBy 'Category' and describe 'Value'
try:
    df_cat = pd.DataFrame({
        "Category": ["A", "A", "B", "B", "B", "C"],
        "Value": [10, 12, 8, 15, 9, 20]
    })
    print("Task 4 - Source data:")
    print(df_cat, "\n")

    grouped = df_cat.groupby('Category')['Value']
    # Apply describe per group
    desc = grouped.describe()
    print("Describe() for 'Value' grouped by 'Category':")
    print(desc, "\n")

    # Short interpretation printed simply:
    for cat, sub in grouped:
        s = sub.describe()
        print(f"Category {cat} -> count={s['count']}, mean={s['mean']:.2f}, std={s['std'] if s['count']>1 else 0:.2f}, min={s['min']}, max={s['max']}")
    print()
except Exception as e:
    print("Error in Task 4:", e)
```

```python
# Concatenation and Merging
# Concatenate df1 & df2 vertically & merge with df3 horizontally
try:
    df1 = pd.DataFrame({"id":[1,2], "val1":[10,20]})
    df2 = pd.DataFrame({"id":[3,4], "val1":[30,40]})
    df3 = pd.DataFrame({"id":[1,2,3,4], "val2":["a","b","c","d"]})

    # Concatenate vertically (rows)
    concat_vertical = pd.concat([df1, df2], axis=0, ignore_index=True)
    print("Task 5 - After vertical concatenation (df1 + df2):")
    print(concat_vertical, "\n")

    # Merge horizontally with df3 using 'id' as key
    merged = pd.merge(concat_vertical, df3, on='id', how='left')
    print("After merging with df3 (on 'id'):")
    print(merged)
    print("Resulting shape:", merged.shape, "\n")
except Exception as e:
    print("Error in Task 5:", e)
```

```python
# Tuples and Sets
# Tuple of fruits and set of numbers attempt to add element to both
try:
    fruits = ("apple", "banana", "cherry")  # tuple (immutable)
    numbers = {1, 2, 3, 4, 5}              # set (mutable)

    print("Task 6 - Original tuple and set:")
    print("tuple:", fruits)
    print("set:", numbers)

    # Attempt to add
    try:
        # Tuples are immutable; you can't append. Demonstrate safe method: create a new tuple.
        fruits_attempt = fruits
        # fruits_attempt.append("orange")  # would raise AttributeError if attempted
        fruits_new = fruits + ("orange",)  # produce a new tuple instead
        print("Added 'orange' by creating new tuple:", fruits_new)
    except Exception as e:
        print("Cannot modify tuple in-place:", type(e).__name__, e)

    # Sets can be modified in-place
    numbers.add(6)
    print("After adding 6 to set:", numbers)

    # Explanation:
    explanation = """
    We cannot change them in-place. We can add() and remove() elements.
    Additionally sets enforce uniqueness and do not preserve order.
    """
    print(explanation)
except Exception as e:
    print("Error in Task 6:", e)
```

```python
# Iterators and Generators
# Iterator class (first 5 even numbers) and generator
try:
    class FirstFiveEvensIterator:
        def __init__(self):
            self._count = 0  # how many yielded so far
            self._next_even = 2

        def __iter__(self):
            return self

        def __next__(self):
            if self._count >= 5:
                raise StopIteration
            val = self._next_even
            self._next_even += 2
            self._count += 1
            return val

    def first_five_evens_generator():
        num = 2
        for _ in range(5):
            yield num
            num += 2

    # Using iterator
    it = FirstFiveEvensIterator()
    print("Task 9 - Iterator produced evens:", list(it))

    # Using generator
    gen = first_five_evens_generator()
    print("Generator produced evens:", list(gen))
    print()
except Exception as e:
    print("Error in Task 9:", e)

```

```python
# Map, Reduce, and Filter
# map(), reduce(), filter()
try:
    from functools import reduce
    nums = [1, 2, 3, 4, 5]

    # map -> square all numbers
    squared = list(map(lambda x: x*x, nums))
    print("Task 10 - Squared (map):", squared)

    # reduce -> product of all numbers
    product = reduce(lambda a, b: a * b, nums, 1)  # initial 1 safe for empty list
    print("Product (reduce):", product)

    # filter -> only even numbers
    evens = list(filter(lambda x: x % 2 == 0, nums))
    print("Evens (filter):", evens, "\n")
except Exception as e:
    print("Error in Task 10:", e)

```

```python
# Object-Oriented Programming
# Rectangle class with area and perimeter
try:
    class Rectangle:
        def __init__(self, length, width):
            if length <= 0 or width <= 0:
                raise ValueError("Length and width must be positive numbers.")
            self.length = length
            self.width = width

        def area(self):
            return self.length * self.width

        def perimeter(self):
            return 2 * (self.length + self.width)

        def __repr__(self):
            return f"Rectangle(length={self.length}, width={self.width})"

    rect1 = Rectangle(4, 6)
    rect2 = Rectangle(5.5, 3)

    print("Task 11 - Rectangle objects and their properties:")
    for r in (rect1, rect2):
        print(f"{r} -> area={r.area()}, perimeter={r.perimeter()}")
    print()
except Exception as e:
    print("Error in Task 11:", e)
```

```python
# Pandas Data Analysis
# Pandas analysis on employees DataFrame
try:
    employees = pd.DataFrame({
        "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
        "Department": ["HR", "IT", "IT", "Finance", "HR"],
        "Salary": [55000, 72000, 64000, 90000, 62000]
    })
    print("Task 12 - Employees DataFrame:")
    print(employees, "\n")

    # Average salary by department
    avg_by_dept = employees.groupby('Department')['Salary'].mean().reset_index().rename(columns={'Salary':'AverageSalary'})
    print("Average salary by department:")
    print(avg_by_dept, "\n")

    # Names of employees with salary > 60000
    high_earners = employees.loc[employees['Salary'] > 60000, 'Name'].tolist()
    print("Employees with salary > 60000:", high_earners, "\n")

    # Add 'Bonus' column: 10% of salary
    employees['Bonus'] = employees['Salary'] * 0.10
    print("Employees DataFrame with Bonus column:")
    print(employees, "\n")

    # Ensure operations were performed on DataFrame correctly
    assert 'Bonus' in employees.columns
    assert employees.loc[employees['Name']=='Bob', 'Bonus'].iloc[0] == 72000 * 0.10
    print("Task 12 verification passed.\n")
except Exception as e:
    print("Error in Task 12:", e)
```
