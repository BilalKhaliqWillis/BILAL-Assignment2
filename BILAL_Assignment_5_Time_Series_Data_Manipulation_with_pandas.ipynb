```python
# Importing the Required Libraries
import pandas as pd
import numpy as np

# Displaying settings for better reading
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)
```

```python
# Loading the Dataset
import pandas as pd

df = pd.read_csv("2016-weather-data-seattle.csv")

df.head()
```

```python
# Clean column names
df.columns = df.columns.str.strip()

print("Columns:", df.columns)

# If Date column exists, we convert it
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
else:
    df.index = pd.to_datetime(df.index)

df.head()
```

```python
# Changing the Frequency from Daily to Weekly Using asfreq()
# Converting daily data to weekly frequency
# 'W' means weekly (Sunday by default)
weekly_df = df.asfreq('W')

# Applying forward fill (padding) to fill missing values
weekly_df = weekly_df.ffill()

# Displaying the result
weekly_df.head()
```

```python
# Extracting Time Components
# Extracting the time components from DatetimeIndex

df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Day_of_Week'] = df.index.day_name()

# Displaying updated dataframe
df.head()
```

```python
# Reindex to Include All Days of 2016
# Creating full date range for 2016
full_date_range = pd.date_range(start='2016-01-01',
                                end='2016-12-31',
                                freq='D')

# Reindexing dataframe
df_reindexed = df.reindex(full_date_range)

# Displaying missing values count
df_reindexed.isna().sum()
```

```python
# Truncating the Dataset to Specific Date Range
# Truncate dataset between Jan 1 and Jan 15, 2016
df_truncated = df.truncate(before='2016-01-01',
                           after='2016-01-15')

df_truncated
```
