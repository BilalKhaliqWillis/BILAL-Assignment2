```python
# Assignment 1: Analyzing Train Arrivals Using Poisson Distribution
# Import Libraries & Set Seed
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson

np.random.seed(42)
```

```python
# Create Initial Observations (20 Hours)
# Number of trains arriving per hour (20 hours)
observations = np.array([
    7, 8, 6, 9, 10,
    8, 7, 6, 9, 11,
    8, 7, 10, 9, 8,
    7, 6, 9, 8, 10
])

# Calculate mean
mean_initial = observations.mean()
mean_initial
```

```python
# Visualizing the Initial Distribution
sns.histplot(observations, kde=True)
plt.xlabel("Number of Trains per Hour")
plt.ylabel("Frequency")
plt.title("Distribution of Train Arrivals-Initial Observations")
plt.show()
```

```python
# Probability Calculations-Initial Mean
thresholds = [9, 10, 11, 12]

for t in thresholds:
    prob = 1 - poisson.cdf(t, mean_initial)
    print(f"P(X > {t}) = {prob:.4f}")
```

```python
# New Observations-Increased Arrivals
new_observations = np.array([
    10, 11, 12, 9, 10,
    11, 12, 10, 9, 11,
    10, 12, 11, 10, 9,
    11, 12, 10, 11, 12
])
```

```python
# Combining the Data & Recalculating Mean
combined_observations = np.concatenate([observations, new_observations])
mean_updated = combined_observations.mean()
mean_updated
```

```python
# Repeating the Probability Calculations-Up to date Mean
for t in thresholds:
    prob = 1 - poisson.cdf(t, mean_updated)
    print(f"P(X > {t}) = {prob:.4f}")
```

```python
# Generating the Synthetic Data of 2000 Samples
synthetic_data = np.random.poisson(mean_updated, size=2000)
```

```python
# Visualizing Synthetic Data
sns.histplot(synthetic_data, kde=True)
plt.xlabel("Number of Trains per Hour")
plt.ylabel("Frequency")
plt.title("Synthetic Train Arrivals Distribution")
plt.show()
```
