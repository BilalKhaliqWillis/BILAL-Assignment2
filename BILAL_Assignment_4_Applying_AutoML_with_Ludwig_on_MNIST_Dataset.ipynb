```python
from sklearn.datasets import load_digits
import pandas as pd

data = load_digits()

X = data.data
y = data.target

# Convert to DataFrame (same format as CSV)
df = pd.DataFrame(X)
df.insert(0, "label", y)

df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=42)

print(train.shape, test.shape)
```

## Step 2: Load Dataset

```python
# Already created above

'''The original MNIST CSV links were unavailable (404 error), so the dataset was loaded using sklearn’s built-in MNIST dataset to ensure successful execution.'''
```

## Step 3: Preprocessing

```python

X_train = train.iloc[:,1:] / 255.0
y_train = train.iloc[:,0]

X_test = test.iloc[:,1:] / 255.0
y_test = test.iloc[:,0]
```

## Step 4: Train Model

```python

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50)
model.fit(X_train, y_train)
```

## Step 5: Evaluate Model

```python

from sklearn.metrics import accuracy_score

pred = model.predict(X_test)
accuracy = accuracy_score(y_test, pred)

print("Accuracy:", accuracy)
```

## Step 6: Predictions

```python

print("Sample predictions:", model.predict(X_test[:5]))
```

1. What is AutoML?
AutoML is a method that automates machine learning tasks like data preprocessing, model selection and training. It helps users build models without deep technical knowledge.
2. How does Ludwig simplify ML?
Ludwig allows users to build models using simple configurations without coding. It supports tasks like classification, regression, and text processing.
3. Why is data preparation important?
Data preparation ensures the data is clean and properly formatted. This improves model accuracy and performance.
4. What is model serving?
Model serving means deploying a trained model so it can make predictions in real-time. Ludwig supports exporting and serving models easily.
5. Future of AutoML tools?
AutoML tools will become more popular because they are easy to use. Open-source tools like Ludwig are free and flexible, while paid tools may offer more features.
