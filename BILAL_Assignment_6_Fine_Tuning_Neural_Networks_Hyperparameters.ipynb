```python
# Importing the Required Libraries
# The Basic Libraries
import numpy as np
import pandas as pd

# Train Test Splitting
from sklearn.model_selection import train_test_split

# Evaluation Metric
from sklearn.metrics import mean_squared_error

# Normalization
from sklearn.preprocessing import StandardScaler

# Deep Learning Libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Loading and Preparing the Dataset
# Loading the dataset
data = pd.read_csv('concrete_data.csv')

# Displaying first few rows
data.head()
```

```python
# Separate Features and Target
# Assuming that the last column is the target -Concrete Strength

X = data.iloc[:, :-1]   # Predictors
y = data.iloc[:, -1]    # Target
```

```python
# BASELINE MODEL
# Building Baseline Model Function
def build_baseline_model(input_dim):
    model = Sequential()

    # One hidden layer with 10 nodes
    model.add(Dense(10, activation='relu', input_dim=input_dim))

    # Output layer -regression -> 1 neuron, no activation
    model.add(Dense(1))

    # Compiling the model
    model.compile(optimizer=Adam(), loss='mean_squared_error')

    return model
```

```python
# Training and Evaluating 50 Times -50 Epochs
mse_list_baseline = []

for i in range(50):

    # Splitting the data (30% test)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3
    )

    # Building model
    model = build_baseline_model(X.shape[1])

    # Training model (50 epochs)
    model.fit(X_train, y_train, epochs=50, verbose=0)

    # Prediction
    y_pred = model.predict(X_test)

    # Compute MSE
    mse = mean_squared_error(y_test, y_pred)

    mse_list_baseline.append(mse)
```

```python
# Report Results
baseline_mean = np.mean(mse_list_baseline)
baseline_std = np.std(mse_list_baseline)

print("Baseline Model Results")
print("Mean of MSE:", baseline_mean)
print("Standard Deviation of MSE:", baseline_std)
```

Discussion
•	The mean MSE displays or shows the average prediction error.
•	The standard deviation shows that how stable the model performance is across different splits.

```python
# NORMALIZED DATA
# Normalizing the Data
scaler = StandardScaler()

mse_list_normalized = []

for i in range(50):

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3
    )

    # Fitting scaler ONLY on training data
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Building the model
    model = build_baseline_model(X.shape[1])

    # Training model
    model.fit(X_train, y_train, epochs=50, verbose=0)

    # Prediction
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mse_list_normalized.append(mse)
```

```python
# Report Results
normalized_mean = np.mean(mse_list_normalized)
normalized_std = np.std(mse_list_normalized)

print("Normalized Model Results")
print("Mean of MSE:", normalized_mean)
print("Standard Deviation of MSE:", normalized_std)
```

Discussion
•	Comparing the normalized_mean with  the baseline_mean.
•	Normalization will reduce MSE notably.
•	Neural networks normaly show better when inputs are on same scales.

•	Expection:- normalized_mean < baseline_mean


```python
# INCREASING EPOCHS -100 Epochs
mse_list_100epochs = []

for i in range(50):

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3
    )

    # Normalization
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    model = build_baseline_model(X.shape[1])

    # Training for 100 epochs
    model.fit(X_train, y_train, epochs=100, verbose=0)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mse_list_100epochs.append(mse)
```

```python
# Report Results
epochs_mean = np.mean(mse_list_100epochs)
epochs_std = np.std(mse_list_100epochs)

print("100 Epochs Model Results")
print("Mean of MSE:", epochs_mean)
print("Standard Deviation of MSE:", epochs_std)
```

Discussion
•	Compareing:- epochs_mean vs normalized_mean
Expectation:-
•	Increasing epochs normally lowers MSE even further
•	This model has more time on hand to learn patterns
•	However too many epochs might cause overfitting

```python
# THREE HIDDEN LAYERS
# Building the Deep Model
def build_deep_model(input_dim):
    model = Sequential()

    model.add(Dense(10, activation='relu', input_dim=input_dim))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))

    model.add(Dense(1))

    model.compile(optimizer=Adam(), loss='mean_squared_error')

    return model
```

```python
# Training Deep Model -50 Epochs, Normalized
mse_list_deep = []

for i in range(50):

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3
    )

    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    model = build_deep_model(X.shape[1])

    model.fit(X_train, y_train, epochs=50, verbose=0)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mse_list_deep.append(mse)
```

```python
# Report Results
deep_mean = np.mean(mse_list_deep)
deep_std = np.std(mse_list_deep)

print("Three Hidden Layers Model Results")
print("Mean of MSE:", deep_mean)
print("Standard Deviation of MSE:", deep_std)
```

```python
# FINAL COMPARISON
print("FINAL COMPARISON")
print("----------------------------------")
print("Baseline Mean MSE:", baseline_mean)
print("Normalized Mean MSE:", normalized_mean)
print("100 Epochs Mean MSE:", epochs_mean)
print("Deep Model Mean MSE:", deep_mean)
```
