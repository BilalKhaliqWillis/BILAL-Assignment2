```python
# Import Libraries
import pandas as pd
import numpy as np

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score

from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.svm import SVC, SVR
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
```

```python
# Upload Files
from google.colab import files
uploaded = files.upload()
```

```python
# Extract Wine ZIP
import zipfile
import os

if os.path.exists("wine+quality.zip"):
    with zipfile.ZipFile("wine+quality.zip", "r") as zip_ref:
        zip_ref.extractall("wine_data")
    print("Wine dataset extracted")
```

```python
# Load Wine Dataset
wine = load_wine()

X_wine = wine.data
y_wine = wine.target

print("Wine shape:", X_wine.shape)
print("Wine classes:", np.unique(y_wine))
```

```python
# Load Car Dataset
cars = pd.read_csv("used_cars.csv")
cars.head()
```

```python
# Clean Price Column
cars['price'] = (
    cars['price']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
# Keep Only Numeric Columns
cars_numeric = cars.select_dtypes(include=[np.number]).dropna()
cars_numeric.head()
```

```python
# Split Features and Target
X_cars = cars_numeric.drop("price", axis=1)
y_cars = cars_numeric["price"]
```

```python
# Feature Scaling
scaler_wine = StandardScaler()
X_wine_scaled = scaler_wine.fit_transform(X_wine)

scaler_cars = StandardScaler()
X_cars_scaled = scaler_cars.fit_transform(X_cars)
```

```python
# TrainTest Split
Xw_train, Xw_test, yw_train, yw_test = train_test_split(
    X_wine_scaled, y_wine, test_size=0.2, random_state=42
)

Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_cars_scaled, y_cars, test_size=0.2, random_state=42
)
```

```python
# Train Classification Models
log_reg = LogisticRegression(max_iter=200)
svc = SVC(C=1.0)
knn_clf = KNeighborsClassifier(n_neighbors=5)

log_reg.fit(Xw_train, yw_train)
svc.fit(Xw_train, yw_train)
knn_clf.fit(Xw_train, yw_train)
```

```python
# Evaluate Classification Models
print("Logistic Regression Accuracy:", accuracy_score(yw_test, log_reg.predict(Xw_test)))
print("SVC Accuracy:", accuracy_score(yw_test, svc.predict(Xw_test)))
print("KNN Accuracy:", accuracy_score(yw_test, knn_clf.predict(Xw_test)))
```

```python
# Train Regression Models
lin_reg = LinearRegression()
svr = SVR(C=100, gamma=0.1)
knn_reg = KNeighborsRegressor(n_neighbors=5)

lin_reg.fit(Xc_train, yc_train)
svr.fit(Xc_train, yc_train)
knn_reg.fit(Xc_train, yc_train)
```

```python
# Evaluate Regression Models
print("Linear Regression RMSE:",
      np.sqrt(mean_squared_error(yc_test, lin_reg.predict(Xc_test))))

print("SVR RMSE:",
      np.sqrt(mean_squared_error(yc_test, svr.predict(Xc_test))))

print("KNN RMSE:",
      np.sqrt(mean_squared_error(yc_test, knn_reg.predict(Xc_test))))
```

```python
 # Hyperparameter Tuning (KNN)
for k in [3, 5, 7, 9]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(Xw_train, yw_train)
    acc = accuracy_score(yw_test, model.predict(Xw_test))
    print(f"KNN (k={k}) Accuracy:", acc)
```
