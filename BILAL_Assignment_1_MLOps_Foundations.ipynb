# Part 1: Conceptual Answers

## 1. What is MLOps?
MLOps (Machine Learning Operations) is a set of practices that combines machine learning, DevOps and data engineering to automate and manage ML systems.

In a Flask ML app, MLOps ensures:
- Model is deployed properly
- Code is version controlled
- Automated testing and deployment

## 2. CI/CD in MLOps
CI/CD automates building, testing, and deploying ML applications.

GitHub Actions helps by:
- Running tests automatically
- Installing dependencies
- Deploying apps

## 3. Cloud Service Integration
Steps:
1. Create Azure DevOps project
2. Connect GitHub repo
3. Configure pipeline

Importance:
- Scalability
- Automation
- Reliability

## 4. ARM Service Connection
It connects Azure DevOps to Azure resources securely.
Used for deploying apps automatically.

## 5. Virtual Environment
It isolates dependencies so projects don't conflict.
Important for reproducibility.

## 6. Pipeline Triggers
Triggers decide when pipelines run (e.g., push to repo).
Defined in YAML file.

## 7. Stages and Jobs
Stages = major steps (build, test, deploy)
Jobs = smaller tasks inside stages

Useful for modular pipelines.

## 8. YAML Pipelines
Benefits:
- Easy to version control
- Reusable

Key elements:
- trigger
- jobs
- steps

# Step 3: Simple Flask App

```python
from flask import Flask, request, jsonify
import numpy as np

app = Flask(__name__)

@app.route('/')
def home():
    return 'ML App Running'

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json(force=True)
    prediction = sum(data['input'])
    return jsonify({'prediction': prediction})

if __name__ == '__main__':
    app.run()
```

# Step 4: requirements.txt

```python
with open('requirements.txt','w') as f:
    f.write('flask\nscikit-learn\npandas')
```

# Step 5: GitHub Actions YAML

```python
yaml_content = '''
name: CI
on: [push]
jobs:
  build:
    runs-on: ubuntu-latest
    steps:
    - uses: actions/checkout@v2
    - name: Set up Python
      uses: actions/setup-python@v2
      with:
        python-version: '3.8'
    - name: Install dependencies
      run: pip install -r requirements.txt
    - name: Run tests
      run: echo 'Tests Passed'
'''

with open('.github_workflow.yml','w') as f:
    f.write(yaml_content)
```

# Step 6: Azure Pipeline YAML

```python
azure_yaml = '''
trigger:
- main

pool:
  vmImage: 'ubuntu-latest'

steps:
- task: UsePythonVersion@0
  inputs:
    versionSpec: '3.8'

- script: pip install -r requirements.txt
- script: echo 'Run Tests'
- script: echo 'Deploy to Azure'
'''

with open('azure-pipelines.yml','w') as f:
    f.write(azure_yaml)
```
